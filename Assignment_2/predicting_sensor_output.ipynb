```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "AirQualityUCI.xlsx"
xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name="AirQualityUCI")

# Convert Date and Time into datetime format
df["Datetime"] = pd.to_datetime(df["Date"].astype(str) + " " + df["Time"].astype(str))
df["Hour"] = df["Datetime"].dt.hour
df["DayOfWeek"] = df["Datetime"].dt.dayofweek
```

```python
# Define input features and targets
features = ["CO(GT)", "NOx(GT)", "NO2(GT)", "T", "RH", "AH", "Hour", "DayOfWeek"]
targets = ["PT08.S1(CO)", "PT08.S2(NMHC)", "PT08.S3(NOx)", "PT08.S4(NO2)", "PT08.S5(O3)"]

X = df[features]
Y = df[targets]
```

```python
# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Normalize data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Model Training and Evaluation
def train_and_evaluate(model, X_train, X_test, Y_train, Y_test, multi_output=False):
    if multi_output:
        predictions = {}
        for target in Y_train.columns:
            model.fit(X_train, Y_train[target])
            predictions[target] = model.predict(X_test)
        Y_pred = pd.DataFrame(predictions, index=Y_test.index)
    else:
        model.fit(X_train, Y_train)
        Y_pred = model.predict(X_test)
    
    mae = mean_absolute_error(Y_test, Y_pred)
    rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
    return mae, rmse
```

```python
# Initialize models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    "MLP Regressor": MLPRegressor(hidden_layer_sizes=(100, 50), activation='relu', solver='adam', max_iter=500, random_state=42)
}
```

```python
# Evaluate all models
results = {}
for name, model in models.items():
    multi_output = name == "Gradient Boosting"
    mae, rmse = train_and_evaluate(model, X_train_scaled, X_test_scaled, Y_train, Y_test, multi_output)
    results[name] = {"MAE": mae, "RMSE": rmse}

# Convert results to DataFrame
results_df = pd.DataFrame(results).T
print(results_df)

```

```python
# Plot results
plt.figure(figsize=(10, 5))
sns.barplot(x=results_df.index, y=results_df["MAE"], color='blue', label="MAE")
sns.barplot(x=results_df.index, y=results_df["RMSE"], color='red', alpha=0.6, label="RMSE")
plt.ylabel("Error")
plt.title("Model Performance Comparison")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
