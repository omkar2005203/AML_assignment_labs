```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
credits.head(5)
```

```python
movies_df = pd.read_csv("tmdb_5000_movies.csv")
movies_df.head(5)
```

```python
# dataset shape

movies_df.shape
```

```python
credits.shape
```

```python
# rename column in credits dataset

new_columns = credits.rename(index=str,columns={"movie_id":"id"})
```

```python
new_columns
```

```python
# merge two dataset
movie_merge_df = movies_df.merge(new_columns,on='id')
```

```python
movie_merge_df.head()
```

```python
movie_merge_df.columns
```

```python
#dropping unecessary columns

cleaned_movies_df = movie_merge_df.drop(columns=['homepage','title_x','title_y','production_countries','status'])
cleaned_movies_df.head(5)
```

![title](img/imdb_formula.png)

```python
cleaned_movies_df.columns
```

```python
# creating variables as per formula
v = cleaned_movies_df['vote_count']
R = cleaned_movies_df['vote_average']
C = cleaned_movies_df['vote_average'].mean()
m = cleaned_movies_df['vote_count'].quantile(0.70)
```

```python
cleaned_movies_df['weighted_average'] = ((v*R)+(m*C))/(v+m)
```

```python
cleaned_movies_df.head()
```

```python
sorted_ranking = cleaned_movies_df.sort_values('weighted_average',ascending=False)
sorted_ranking[['original_title','vote_average', 'vote_count','weighted_average','popularity']].head(20)
```

```python
##plotting graph

weight_avg = cleaned_movies_df.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_avg['weighted_average'].head(10),y=weight_avg['original_title'].head(10),palette='pastel')
plt.xlim(4,10)
plt.title('Recommended Movies')
plt.xlabel('Weighted average')
plt.ylabel('Movie titles')
```

```python
scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(cleaned_movies_df[['weighted_average','popularity']])
normalized_movie_df = pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
normalized_movie_df.head()
```

```python
cleaned_movies_df[['normalized_weighted_average','normalized_popularity']] = normalized_movie_df
```

```python
cleaned_movies_df.head()
```

```python
cleaned_movies_df['score'] = cleaned_movies_df['normalized_weighted_average']*0.5 + cleaned_movies_df['popularity']*0.5
```

```python
score_df = cleaned_movies_df.sort_values(['score'],ascending=False)
score_df.head()
```
